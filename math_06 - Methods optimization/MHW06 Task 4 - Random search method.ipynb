# MHW06 Task 4 - Random search method

```python
import numpy as np
import numpy.random as rand
import sympy as sp
```

```python
# Input function for : f(x) = 4(x1 - 5)**2 - (x2 - 6)**2

# Input symbols and symbolyc funtion for X variable
x1, x2 = sp.symbols('x1, x2')
symbol_function = 4*(x1 - 5)**2 + (x2**0.5 - 6)

# Initail  values

# Initail X point
x_inital = np.array([8,9])

# Initial step and mininal value of the step
step = 1
step_min = 0.8
# Maximum number of failed tests current iteration
M  = 3
# Accuracy of method
alpha = 1.6
betta = 0.5
max_iterations = 10
```

```python
# functions

def rand_fun(values, step):
    '''substitute 2 numerical arguments in symbolic for functions'''
    global ksi
    norm_ksi = np.linalg.norm(ksi.astype(float))
    return np.array( values + step * (ksi / norm_ksi), dtype=float)

def f_num(values):
    '''substitute 2 numerical arguments in symbolic for functions'''
    global symbol_function
    return np.array(symbol_function.subs([(x1,values[0]),(x2,values[1])]), dtype=float)

def f_direction(values_x, values_y):
    '''Function to check direction of random choises'''
    global alpha
    return values_x + alpha * (values_y - values_x)

def new_step_value(step):
    '''New value of step'''
    global alpha
    return alpha * step
```

```python
# Main alhorytm

x_inital = np.array([8,9])
x_eval = x_inital
step_eval = step

print('\n input points: ',x_eval,'\n')

for index in range(max_iterations):

    print('\n\n iterations:', index + 1)

    # Searching if random field x and f(x) values
    for iteration in range(M):
    
        # Random x vector
        ksi = np.random.uniform(-1,1,2)
    
        if (f_num(rand_fun(x_eval, step)) < f_num(x_eval)):
  
            # Evaluation function random vector
            value_y = f_num(rand_fun(x_eval, step_eval))
           
            print(' M: ',iteration +1)
            print(' ksi = ',ksi)
            print(' y(random) = ',value_y)
            print(' f(x) = ',f_num(x_eval))
            print(' x =',x_eval)
               
            
            break
            
    #print( f_num(f_direction(x_eval, value_y)),'***',f_num(x_eval))
    
    # Cheking direction of evaluation
    if f_num(f_direction(x_eval, value_y)) < f_num(x_eval):
        # Step modify and new X value
        step_eval = new_step_value(step_eval)
        x_eval = f_direction(x_eval, value_y)
        # output iteration result
        print('\n new step: ',step_eval)
        print(' new x =',x_eval)

    # Modify step in random serch field of x values   
    if (iteration == M) and (step_eval > step_min):
        step_eval = betta * step_eval
        print('\n modify step: ',step_eval)   
    elif step_eval < step_min:
        print('\n\n---->',step_eval,'<',step_min,'<----- stop this ^^')
        break
```
