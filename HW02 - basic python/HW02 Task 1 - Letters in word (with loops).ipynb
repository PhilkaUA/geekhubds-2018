# ДЗ 2:
1. Дана строка, содержащая текст со знаками препинания. Слова разделены пробелом. После знаков препинания пробел стоит всегда. Необходимо разбить строку на слова.
- уровень 1: достаточно просто получить список слов, на которые можно разделить строку
- уровень 2: удалить из этого списка все знаки препинания
- уровень 3: оставить в списке только уникальные слова без знаков препинания (Слово, в котором первая буква большая и такое же слово, в котором первая буква маленькая считаем одинаковым: Все=все)
Пример строки:
"Все йде, все минає і краю немає , \n Куди ж воно ділось? відкіля взялось! \n І дурень, і мудрий нічого не знає. \n Живе... умирає... одно зацвіло , \n А друге зав’яло, навіки зав’яло... \n І листя пожовкле вітри рознесли."
2. В чем разница между hashable и immutable

# Уровень 1 команда .split

```python
# Вводим текст со знаками препинания
text = "Все йде, все минає і краю немає, \n Куди ж воно ділось? відкіля взялось!\n І дурень, і мудрий нічого не знає. \n Живе... умирає... одно зацвіло, \n А друге зав’яло, навіки зав’яло... \n І листя пожовкле вітри рознесли."
```

```python
# Выделяем слова с текса
slova = text.split()
```

```python
# Вывод результата
print(slova)
```

# Уровень 2 команды: len, append, for

```python
# Считаем слова
slova_length = len(slova)
# Cоздаем список без ,
slova_no_koma = []
# Выделяем со слов ,
for i in range(slova_length):
    slova_no_koma.append(slova[i].split(',')[0])
# Вывод результата
print(slova_no_koma)
```

```python
# Cоздаем список для без . и ...
slova_no_tochk = []
# Выделяем со слов . и ...
for i in range(slova_length):
    slova_no_tochk.append(slova_no_koma[i].split('.')[0])
# Вывод результата
print(slova_no_tochk)
```

```python
# Cоздаем список для без ?
slova_no_vopros = []
# Выделяем со слов ?
for i in range(slova_length):
    slova_no_vopros.append(slova_no_tochk[i].split('?')[0])
# Вывод результата
print(slova_no_vopros)
```

```python
# Cоздаем список для без !
slova_no_voskl = []
# Выделяем со слов !
for i in range(slova_length):
    slova_no_voskl.append(slova_no_vopros[i].split('!')[0])
# Вывод результата
print(slova_no_voskl)
```

# Уровень 3 (способ 1) команды if, for, set, lower

```python
# Cоздаем cписок для слов без ,
slova_no_koma_ = []
# Выделяем слова с ,
for i in range(slova_length):
    if len(slova[i].split(',')) == 1:
        slova_no_koma_.append(slova[i].split(',')[0])
# Вывод результата
print(slova_no_koma_)
```

```python
# Cоздаем cписок для слов без . и ...
slova_no_tochk_ = []
# Выделяем слова с . и ...
for i in range(len(slova_no_koma_)):
    if len(slova_no_koma_[i].split('.')) == 1:
        slova_no_tochk_.append(slova_no_koma_[i].split('.')[0])
# Вывод результата
print(slova_no_tochk_)
```

```python
# Cоздаем  cписок для слов без ?
slova_no_vopros_ = []
# Выделяем  слова с ?
for i in range(len(slova_no_tochk_)):
    if len(slova_no_tochk_[i].split('?')) == 1:
        slova_no_vopros_.append(slova_no_tochk_[i].split('?')[0])
# Вывод результата
print(slova_no_vopros_)
```

```python
# Cоздаем  cписок для слов без !
slova_no_voskl_ = []
# Выделяем  слова с ?
for i in range(len(slova_no_vopros_)):
    if len(slova_no_vopros_[i].split('!')) == 1:
        slova_no_voskl_.append(slova_no_vopros_[i].split('!')[0])
# Вывод результата
print(slova_no_voskl_)
```

```python
# Преобразовываем регистр
result = [];
for i in range(len(slova_no_voskl_)):
    result.append(slova_no_voskl_[i].lower())
# Вывод результата
print(result)
```

```python
# Выделяем уникальные слова
result_all = set(result)
# Вывод результата
print(result_all)
```

# Уровень 3 (способ 2) через буквы

```python
# Разделяем все слова на символы - слова лежат в отдельных списках
slova_all = [];
for i in range(slova_length):
    slova_tmp = [];
    slova_tmp += slova[i]
    slova_all.append(slova_tmp)
print(slova_all)
```

```python
# Отделяем слова с ,
slova_no_koma_n = [];
for j in range(len(slova_all)):
    if slova_all[j][-1] != ",":
        slova_no_koma_n.append(slova_all[j])  
print(slova_no_koma_n)
```

```python
# Отделяем слова с . и ...
slova_no_tochka_n = [];
for j in range(len(slova_no_koma_n)):
    if slova_no_koma_n[j][-1] != ".":
        slova_no_tochka_n.append(slova_no_koma_n[j])  
print(slova_no_tochka_n)
```

```python
# Отделяем слова с ?
slova_no_vopros_n = [];
for j in range(len(slova_no_tochka_n)):
    if slova_no_tochka_n[j][-1] != "?":
        slova_no_vopros_n.append(slova_no_tochka_n[j])  
print(slova_no_vopros_n)
```

```python
# Отделяем слова с !
slova_no_voskl_n = [];
for j in range(len(slova_no_vopros_n)):
    if slova_no_vopros_n[j][-1] != "!":
        slova_no_voskl_n.append(slova_no_vopros_n[j])  
print(slova_no_voskl_n)
```

```python
# Обьеденяем буквы в слова
slova_all_n = [];
for j in range(len(slova_no_voskl_n)):
    tmp_i = '';
    slova_all_tmp = [];
    for i in range(len(slova_no_voskl_n[j])):
        tmp_i = tmp_i + slova_no_voskl_n[j][i]
        slova_all_tmp.append(tmp_i)
    slova_all_n.append(slova_all_tmp[-1])
# Выводим результат
print(slova_all_n)
```

```python
# Преобразовываем регистр
result_n = [];
for i in range(len(slova_all_n)):
    result_n.append(slova_all_n[i].lower())
# Вывод результата
print(result_n)
```

```python
# Выделяем уникальные слова
result_all_n = set(result_n)
# Вывод результата
print(result_all_n)
```

# Уровень 4

убрать скобки и кавычки, которые примыкают к слову спереди или сзади (вместо скобок убрал тире)

```python
# Входной текст
text2 = "“Здоров, Еоле, пане-свату! Ой, як ся маєш, як живеш? — Сказала, як ввійшла у хату, Юнона. — Чи гостей ти ждеш?”"
```

```python
# Выделяем слова с текса
slova2 = text2.split()
# Выводим текст по словам
print(slova2)
```

```python
# Записываем слова и удаляем левые "
slova_kavychky_l = [];
for i in range(len(slova2)):
    slova_kavychky_l.append(slova2[i].lstrip("“"))
print(slova_kavychky_l)
```

```python
# Записываем слова и удаляем правые "
slova_kavychky_r = [];
for i in range(len(slova_kavychky_l)):
    slova_kavychky_r.append(slova_kavychky_l[i].rstrip("”"))
print(slova_kavychky_r)   
```

```python
# Выделяем слова с дефисом
slova_tire = []
for i in range(len(slova_kavychky_r)):
    slova_tire.append(slova_kavychky_r[i].split('-'))
# Вывод результата
print(slova_tire)
```

```python
# Обьеденяем буквы в слова
slova_all_last = [];
for j in range(len(slova_tire)):
    tmp_i = '';
    slova_all_tmp = [];
    for i in range(len(slova_tire[j])):
        tmp_i = tmp_i + slova_tire[j][i]
        slova_all_tmp.append(tmp_i)
    slova_all_last.append(slova_all_tmp[-1])
# Выводим результат
print(slova_all_last)
```
